```python
import matplotlib.pyplot as plt
```

```python
#Bar plot of Arrests by Race
arrests_by_race = data.groupby('driver_race')['is_arrested'].sum()
plt.figure(figsize=(10, 6))
arrests_by_race.plot(kind='bar', color='orange')
plt.title('Arrests by Race')
plt.xlabel('Race')
plt.ylabel('Number of Arrests')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['driver_age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Driver Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Bar plot of Violation Types
plt.figure(figsize=(10, 6))
data['violation'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Violation Types')
plt.xlabel('Violation')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
